```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
def plot_runtime(input):
    data_plane_sweep_csv = 'runtime-tester-output-{}-plane-sweep.csv'.format(input)
    data_brute_force_csv = 'runtime-tester-output-{}-brute-force.csv'.format(input)

    data_sweep = pd.read_csv(data_plane_sweep_csv)
    data_sweep['segments'] = data_sweep['vertical segments'] + data_sweep['horizontal segments']
    data_sweep.sort_values(by=['segments'], inplace=True)
    
    data_brute = pd.read_csv(data_brute_force_csv)
    data_brute['segments'] = data_brute['vertical segments'] + data_brute['horizontal segments']
    data_brute.sort_values(by=['segments'], inplace=True)

    sns.set()

    # t = np.arange(1,26)
    # s = x * t + y

    plt.plot(data_sweep['segments'],data_sweep['runtime(ms)'], label='plane sweep')
    plt.plot(data_brute['segments'],data_brute['runtime(ms)'], label='brute force')
    # plt.plot(t, s, label='approximation')
    plt.xlabel('number of segments')
    plt.ylabel('execution time (ms)')

    plt.legend()

    plt.savefig('runtime-plot-{}.png'.format(input))
```

```python
plot_runtime('1horizontal')
```

```python
plot_runtime('1vertical')
```

```python
plot_runtime('noOverlap')
```

```python
plot_runtime('overlap')
```

```python
plot_runtime('ranGenerater')
```
